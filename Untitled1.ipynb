```python
import matplotlib.pyplot as plt
x=['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
y1=[13,16,15,18,16,17,16]
y2=[17,14,17,16,15,15,14]
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('Daily')                                 #x축 이름
plt.ylabel('no. of Books Sold')                     #y축 이름
plt.legend(['Sold','On Shalves'],loc='lower left')  #범례
plt.title('Le Petit Prince : Sold & Shalves')       #그래프 제목
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y,'mo')
plt.axis([0,6,0,30])
plt.show()
```
